```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston

data = load_boston()
train = data.data
train = pd.DataFrame(train,
            columns=data['feature_names'])
target = data.target

num_cols = train.shape[1]
```

## One-hot-Encoding

```python
train_oe = pd.get_dummies(train, columns=['CHAS'])
train_oe
```

## Label-Encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(train['CHAS'])
train_le = train.copy()
train_le['CHAS'] = le.transform(train['CHAS'])
train_le
```

## Feature-Hashing

```python
from sklearn.feature_extraction import FeatureHasher

fh = FeatureHasher(n_features=5, input_type='string')
hash_train = fh.transform(train['RAD'].astype(str).values)
hash_train = pd.DataFrame(hash_train.todense(), columns=[f'RAD_{i}' for i in range(5)])
train_fh = pd.concat([train, hash_train], axis=1)
train_fh.drop(['RAD'], axis=1, inplace=True)
train_fh
```

## Frequency_Encoding

```python
freq = train['RAD'].value_counts
train_fre = train.copy()
train_fre['RAD'] = train_fre['RAD'].map(freq)
train_fre
```

## Target-Encoding

```python
from sklearn.model_selection import KFold

col_name = 'RAD'
data_tmp = pd.DataFrame({col_name: train[col_name], 'target': target})
target_mean = data_tmp.groupby(col_name).mean()

tmp = np.repeat(np.nan, train.shape[0])

kf = KFold(n_splits=4, shuffle=True, random_state=72)
for idx_1, idx_2 in kf.split(train):
    target_mean = data_tmp.iloc[idx_1].groupby(col_name)['target'].mean()
    tmp[idx_2] = train[col_name].iloc[idx_2].map(target_mean)
    train_te = train.copy()
    train_te['RAD'] = tmp
    
train_te
```
